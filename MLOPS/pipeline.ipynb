```python
import pandas as pd
import numpy as np
import joblib
import warnings
import gdown


from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
url = 'https://drive.google.com/uc?id=13cbr0Jghx2oXZuSprzJhWeHEwDcgxxBT'
output = 'weatherAUS.csv'
gdown.download(url, output, quiet=False)
warnings.simplefilter('ignore')
```

```python
file_path= 'weatherAUS.csv'
df = pd.read_csv(file_path, sep=',',engine='python')
```

MLOPS

```python
from regression_class import feat_eng

df = feat_eng(df)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df,
                                              df['RainfallTomorrow'].values.reshape(-1,1),
                                              test_size=0.1,
                                              random_state=12,
                                             )

X_train.drop(['RainfallTomorrow', 'RainTomorrow_Yes'], axis=1, inplace=True)
X_test.drop(['RainfallTomorrow', 'RainTomorrow_Yes'], axis=1, inplace=True)
```

```python
from regression_class import NeuralNetworkRegressor

pipeline = Pipeline([
  # ('Feature Engineering', FunctionTransformer(feat_eng)),
  ('inputer', SimpleImputer(strategy='mean')),
  ('scaler', StandardScaler()),
  ('model', NeuralNetworkRegressor())
])

pipeline.fit(X_train, y_train)

joblib.dump(pipeline, 'regression_pipeline.joblib')
```

```python
X_train_clas, X_test_clas, y_train_clas, y_test_clas = train_test_split(df,
                                              df['RainTomorrow_Yes'].values.reshape(-1,1),
                                              test_size=0.1,
                                              random_state=12,
                                             )

X_train_clas.drop(['RainTomorrow_Yes', 'RainfallTomorrow'], axis=1, inplace=True)
X_test_clas.drop(['RainTomorrow_Yes', 'RainfallTomorrow'], axis=1, inplace=True)
```

```python
from classification_class import NeuralNetworkClassifier

pipeline = Pipeline([
  # ('Feature Engineering', FunctionTransformer(feat_eng)),
  ('inputer', SimpleImputer(strategy='mean')),
  ('scaler', StandardScaler()),
  ('model', NeuralNetworkClassifier())
])

pipeline.fit(X_train_clas, y_train_clas)

joblib.dump(pipeline, 'classification_pipeline.joblib')
```
